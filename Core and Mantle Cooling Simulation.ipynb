```python
import numpy as np
import matplotlib.pyplot as plt

# Parameters
initial_temperature = 5000  # Initial temperature in °C
cooling_rate = 0.01  # Cooling rate in °C/year
time_period = 1e9  # Time period in years

def core_mantle_cooling(initial_temperature, cooling_rate, time_period):
    years = np.linspace(0, time_period, 500)
    temperatures = initial_temperature - cooling_rate * years
    return years, temperatures

# Simulate core and mantle cooling
years, temperatures = core_mantle_cooling(initial_temperature, cooling_rate, time_period)

# Plotting the cooling curve
plt.figure(figsize=(10, 5))
plt.plot(years, temperatures, color='r')
plt.xlabel('Time (years)')
plt.ylabel('Temperature (°C)')
plt.title('Core and Mantle Cooling Over Time')
plt.show()
```
